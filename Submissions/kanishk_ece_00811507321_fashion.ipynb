```python
import numpy as np
np.set_printoptions(threshold=np.inf)
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.compose import ColumnTransformer  
#Importing the dataset
dataset = pd.read_csv('/content/drive/MyDrive/data set/Churn_Modelling.csv')
dataset.head()
```

```python
# dataset = pd.read_csv('Churn_Modeling.csv')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
sklearn.preprocessing import StandardScaler
X_scale = StandardScaler()
X_train[:, [0, 3, from4, 5, 6, 9]] = X_scale.fit_transform(X_train[:, [0, 3, 4, 5, 6, 9]])
X_test[:, [0, 3, 4, 5, 6, 9]] = X_scale.transform(X_test[:, [0, 3, 4, 5, 6, 9]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
geography_encoder = OneHotEncoder(dtype=int, drop='first')
gender_encoder = OneHotEncoder(dtype=int, drop='first')
ct = ColumnTransformer([('geography', geography_encoder, [1]), ('gender', gender_encoder, [2])], remainder='passthrough')

X_train = ct.fit_transform(X_train)
X_test = ct.fit_transform(X_test)
```

```python
X_train = np.asarray(X_train).astype(np.float32)
X_test = np.asarray(X_test).astype(np.float32)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import L2

model = Sequential()
```

```python
regularizer = L2(0.01)
```

```python
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=32, activation='relu'))
```

```python
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=20)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score

predictions = model.predict(X_test)
predictions = (predictions > 0.5)

cm = confusion_matrix(y_test, predictions)
accuracy = accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```
